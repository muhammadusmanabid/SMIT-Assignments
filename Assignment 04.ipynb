## Question 02

```python
class Rectangle():
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def rec_area(self):
        print(f"Length: {self.length}  Width: {self.width}.\n")
        return f"Area: {self.length * self.width}."

    def rec_perimeter(self):
        print(f"Length: {self.length}  Width: {self.width}.")
        return f"Perimeter: {2 * (self.length + self.width)}."

R1 = Rectangle(12.3, 33.33)
R2 = Rectangle(14.5, 35.43)

print(f"Rectangle 1: {R1.rec_area()}\n")
print(f"Rectangle 1: {R1.rec_perimeter()}\n")

print(f"Rectangle 2: {R2.rec_area()}\n")
print(f"Rectangle 2: {R2.rec_perimeter()}")
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_details(self):
        return f"Make by {self.make} and the model is {self.model} in year {self.year}"

class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def display_details(self):
        prev_method = super().display_details()
        print(f"{prev_method} and the number of doors are {self.doors}")

Car1 = Car("Toyota",  "Corolla", "2025", "4")
Car1.display_details()
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount <= 0:
            return False
        elif amount % 500 == 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"You withdrew amount Rs.{amount} successfully!")
            else:
                return False
        else:
            return f"Invalid amount entered."

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Amount of Rs.{amount} depsoit successfully!")
            print(f"Your updated balance is {self.balance}.")
        else:
            return "Invalid amount entered."

B1 = BankAccount("2345 6789 1265", 10000)
B1.withdraw(5000)
```

## Question 05

```python
import math

class Shape:
    def area_calculation(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self. radius = radius

    def area_calculation(self):
        area = round(math.pi * (self.radius ** 2), 2)
        return f"Area of circle is: {area}"

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area_calculation(self):
        area = round(0.5 * self.base * self.height)
        return f"Area of triangle is: {area}."

C1 = Circle(12.5)
T1 = Triangle(12.45, 15.5)

print(C1.area_calculation())
print(T1.area_calculation())
```

## Question 06

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calc_ann_salary(self):
        ann_sal = 12 * self.salary
        return f"{self.name} annual salary is {ann_sal}."

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def calc_ann_salary(self):
        ann_bon_sal = 12 * self.salary * 0.30
        prev = super().calc_ann_salary()
        return f"{prev} After bonus annual salary is {ann_bon_sal}."

M1 = Manager("Muhammad Usman", 150000, "SQA")

M1.calc_ann_salary()
    
```

## Question 09

```python
class Bank:
    def __init__(self, bank_names, branches):
        self.bank_names = bank_names
        self.branches = []

    def add_branch(self, branch_name):
        if branch_name not in self.branches:
            self.branches.append(branch_name)
            return f"Branch {branch_name} successfully added!"
        else:
            return f"Branch {branch_name} already exists."
            
    def remove_branch(self, branch_name):
        if branch_name in self.branches:
            self.branches.remove(branch_name)
            return f"Branch {branch_name} removed successfully!"
        else:
            return f"Branch {branch_name} not found."

    def display_branches(self):
        return f"Branches: {self.branches}"

B1 = Bank(bank_names = "Bank Al Falah", branches = [])

while True:
    print("1. Add Branch")
    print("2. Remove Branch")
    print("3. Display All Branches")
    print("4. Exit")

    choice = input("Enter you choice: ")

    if choice == "1":
        branch_name = input("Enter branch name: ")
        print(B1.add_branch(branch_name))

    elif choice == "2":
        branch_name = input("Enter branch name: ")
        print(B1.remove_branch(branch_name))

    elif choice == "3":
        print(B1.display_branches())

    elif choice == "4":
        break
```

## Question 11

```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance = 0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return False
        else:
            self.balance += amount
            print(f"Amount of Rs.{amount} is successfully added!")
            return f"Your current balance is: {self.balance}."

    def withdrawal(self, amount):
        if amount <= 0:
            return False
        elif amount <= self.balance and amount % 500 == 0:
            self.balance -= amount
            print(f"You withdrew amount of Rs.{amount} successfully!")
            return f"Your current balance is of Rs.{self.balance}."
        else:
            return "Invalid Amount entered."

    def transfer(self, other_account, amount):
        if amount <= self.balance:
            self.balance -= amount
            other_account.balance += amount
            return f"You transfer amount of Rs.{amount} to {other_account.account_holder} successfully!"
        else:
            return False

    def display(self):
        return f"Balance: {self.balance}"

Person1 = BankAccount("432156789014", "Muhammad Usman", 50000)
Person2 = BankAccount("9014 5678 4321", "Syed Anas Raza", 20000)

Person1.transfer(Person2, 20000)
```

```python
class University:
    def __init__(self, name):
        self.name = name
        self.departments = []

    def add_department(self, dept_name):
        if dept_name not in self.departments:
            self.departments.append(dept_name)
            return f"The department {dept_name} ssuccessfully added!"
        else:
            return f"Department {dept_name} already exists."

    def remove_department(self, dept_name):
        if dept_name in self.departments:
            self.departments.remove(dept_name)
            return f"Department {dept_name} removed successfully!"
        else:
            return f"Department {dept_name} not found."

    def display(self):
        return f"Departments: {self.departments}"

U1 = University("Indus")

print(U1.add_department("FCIT"))
print(U1.add_department("FCD"))

print(U1.remove_department("FCD"))

print(U1.display())
```
