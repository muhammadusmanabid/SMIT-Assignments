```python
import re

def remove_titles(name):
    name = name.strip().lower()
    
    # Remove prefix titles
    name = re.sub(r'^(mr|mrs|ms|dr|miss)\.\s*', '', name)
    
    # Remove suffix titles
    name = re.sub(r'\s+(md|dds|dvm|phd|jr\.?|iii)$', '', name)
    
    return name.strip()

df["student_name"] = df["student_name"].apply(remove_titles)
```

## Handling City Column

```python
df["city"].unique()
```

```python
df["city"].astype(str).str.lower().str.strip()
```

```python
def remove(name):
    if pd.isna(name):
        return np.nan

    name = str(name).strip().lower()

    if name == "0":
        return np.nan

    if "123" in name:
        parts = name.split("123")
        if len(parts) > 1:
            return parts[1]

    return name

df["city"] = df["city"].apply(remove)
```

```python
import numpy as np

# Cities to fill
cities_to_fill = ["hudsonborough", "west lancebury", "new samuelfort"]

# Unwanted values
unwanted = ["Multan", "Peshawar", "Quetta"]

# Mask: rows jahan value unwanted, number, or empty string
mask = df["city"].apply(lambda x: str(x).strip().lower() in unwanted or str(x).isdigit() or str(x).strip() == "")

# Fill those rows with repeated cities_to_fill
df.loc[mask, "city"] = np.resize(cities_to_fill, mask.sum())
```

```python
# Unwanted cities
unwanted = ["Multan", "Peshawar", "Quetta"]

# Replace with "hudsonborough"
df.loc[df["city"].isin(unwanted), "city"] = "hudsonborough"
```
