## Write a function is_strong(password) that checks if the password:

* Has at least 8 characters

* Contains numbers

* Contains uppercase letters

* Contains lowercase letters

### Return True or False.

```python
def is_strong(password):
    if len(password) < 8:
        return False

    has_upper = False
    has_lower = False
    has_numbers = False
    
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_numbers = True

    return has_upper, has_lower, has_numbers

password = input("Enter password (must contain UpperCase,LowerCase,Numbers,8 digit): ")
print(is_strong(password))
```

## Write a function analyze_results(marks_list)

    Highest marks

    Lowest marks

```python
def analyze_results(marks_list):
    highest_marks = max(marks_list)
    lowest_marks = min(marks_list)
    average_marks = sum(marks_list) / len(marks_list)
    pass_count = sum(1 for mark in marks_list if mark >= 50)
    pass_ratio = (pass_count / len(marks_list)) * 100

    results = {
        "Highest Marks" : highest_marks,
        "Lowest Marks" : lowest_marks,
        "Average Marks" : average_marks,
        "Pass Count" : pass_count,
        "Pass Ratio" : pass_ratio
    }

    return f"Results: {results}"
marks_list = []

for i in range(5):
    marks = int(input(f"Enter marks for student {i + 1}"))
    marks_list.append(marks)

print(analyze_results(marks_list))
```

## Airline Seat Booking With Constraints

### Write a function book_seat(seats, seat_requested):

### Constraints:

    If seat_requested not in seats ‚Üí return "Invalid Seat"
    
    If seat is "booked" ‚Üí return "Seat Already Booked"
    
    Otherwise book the seat and return updated seats dictionary

### Example seats dictionary:
{
  "1A": "free",
  "1B": "booked",
  "2A": "free"
}

```python
def book_seat(seats, seat_requested):
    if seat_requested not in seats:
        return "Invalid Seat!"

    if seats[seat_requested] == "booked":
        return "Seat already booked!"

    seats[seat_requested] = "booked"
    return seats

seats = {
    "1A" : "free",
    "2A" : "booked",
    "3A" : "free",
    "4A" : "free",
    "5A" : "booked"
}

seat_requested = input("Enter seat number you want to book: ")
print(book_seat(seats, seat_requested))
```

### Calculate the subtotal (price √ó qty for each item)

    Add 8% tax
    
    Add 5% service charge
    
    Return the final bill
    
    Also print each item in this format:
    
    ItemName ‚Äî Qty: X ‚Äî Total: Y

### üìù Example Input:
    items = [
        {"name": "Burger", "price": 450, "qty": 2},
        {"name": "Fries", "price": 200, "qty": 1},

## Create a Python program that:

### Defines a function calculate_patient_bill(patient)

    patient is a dictionary with keys: "name", "days_admitted", "room_type", "tests"

### Rules:

    Room charges per day:
    
        "General": 2000
        
        "Semi-Private": 4000
        
        "Private": 6000
        
        Test charges = sum of all test costs in tests list
        
        Apply 10% insurance discount on total
        
        Print the full bill with patient name, room charges, test charges, discount, and net bill

### üìù Example Input:
    patient = {
        "name": "Ali",
        "days_admitted": 3,
        "room_type": "Private",
        "tests": [1500, 2000, 500]
    }

### üìù Example Output:
    Patient: Ali
    Room Charges: 18000
    Test Charges: 4000
    Subtotal: 22000
    Insurance Discount (10%): 2200
    Net Bill: 19800

```python
def calculate_patient_bill(patient):
    name = patient["name"]
    days_admitted = patient["days_admitted"]
    room_type = patient["room_type"]
    tests = patient["tests"]
    charges = 0
    
    if room_type == "General":
        charges += 2000
    elif room_type == "Semi Private":
        charges += 4000
    elif room_type == "Private":
        charges += 6000
    else:
        charges += 0

    room_charges = charges * days_admitted
    test_total = sum(tests)
    total_amount = room_charges + test_total
    insurance_discount = total_amount * 0.10
    final_bill = total_amount - insurance_discount

    print("------------------------------------------")
    print(f"Patient: {name}")
    print(f"Room Charges: {room_charges}")
    print(f"Test Charges: {test_total}")
    print(f"Subtotal: {total_amount}")
    print(f"Insurance Discount (10%): {insurance_discount}")
    print(f"Net Bill: {final_bill}")

name = input("Enter your name: ")
days_admitted = int(input("Enter days you've been admitted: "))
room_type = input("Enter your room type: ")

tests = []

num_tests = int(input("How many types have been done: "))

for i in range(num_tests):
    amount = int(input(f"Enter amount of {i + 1} test: "))
    tests.append(amount)

patient = {
    "name" : name,
    "days_admitted" : days_admitted,
    "room_type" : room_type,
    "tests" : tests
}

calculate_patient_bill(patient)
```

### Defines a function course_progress(student)

    student is a dictionary: "name", "lessons_completed", "total_lessons", "quizzes" (list of scores)

### The function should:

    Calculate lesson completion %
    
    Calculate average quiz score
    
    Determine progress status:
    
    ‚â• 80% completion ‚Üí "On Track"
    
    50‚Äì79% ‚Üí "Needs Improvement"
    
    <50% ‚Üí "Behind"
    
    Print a progress report

### üìù Example Input:
    student = {
        "name": "Ubaid",
        "lessons_completed": 8,
        "total_lessons": 10,
        "quizzes": [80, 75, 90]
    }

### üìù Example Output:
    Student: Ubaid
    Lesson Completion: 80%
    Average Quiz Score: 81.7
    Progress Status: On Track

```python
def course_progress(student):
    name = student["name"]
    lessons_completed = student["lessons_completed"]
    total_lessons = student["total_lessons"]
    quizzes = student["quizzes"]

    lesson_completion = lessons_completed / total_lessons * 100
    average_quiz_score = sum(quizzes) / len(quizzes)

    if lesson_completion >= 80:
        status = "On Track!"
    elif 50 <= lesson_completion <= 79:
        status = "Needs Improvement!"
    else:
        status = "Behind"

    print("--------------------------")
    print(f"Student: {name}")
    print(f"Lesson Completion: {lesson_completion}")
    print(f"Average Quiz Score: {average_quiz_score}")
    print(f"Progress Status: {status}")

name = input("Enter your name: ")
lessons_completed = int(input("Enter lessons you've completed: "))
total_lessons = int(input("Enter total number of lessons: "))

quizzes = []

no_of_quizzes = int(input("Enter no.of quizzes has been assessed: "))

for i in range(no_of_quizzes):
    quiz_mark = int(input(f"Enter marks for quiz {i + 1}: "))
    quizzes.append(quiz_mark)

student = {
    "name" : name,
    "lessons_completed" : lessons_completed,
    "total_lessons" : total_lessons,
    "quizzes" : quizzes
}

course_progress(student)
```

## An online store calculates final prices dynamically:

### Function calculate_discount(price, category) ‚Üí category-based discounts:

    Electronics > $1000 ‚Üí 15%
    
    Clothing > $500 ‚Üí 10%
    
    Groceries ‚Üí 5%
    
    Function apply_coupon(price, coupon_code) ‚Üí nested function applies coupon:
    
    SAVE10 ‚Üí $10 off
    
    SAVE20 ‚Üí $20 off if price > $200
    
    Function final_price(price, category, coupon) ‚Üí combines discount + coupon + tax (8%)

### üì• Input Example:
    final_price(1200, "Electronics", "SAVE10")

### üìù Example Output:
    Price after discount: 1020
    Price after coupon: 1010
    Final Price with Tax: 1090.8

```python
def category_discount(price, category):
    category = category.lower()

    if category == "electronics":
        if price > 1000:
            discount = price * 0.15
            final_price = price-discount
        else:
            return "Price must be above 1000 PKR"

    elif category == "clothing":
        if price > 500:
            discount = price * 0.10
            final_price = price - discount
        else:
            return "Price must be above 500 PKR"

    elif category == "groccery":
        discount = price * 0.05
        final_price = price - discount

    else:
        return "Invalid category"

    print(f"Price after category discount: {final_price:.2f}")
    return final_price

def apply_coupon(price, coupon_code):
    coupon_code = coupon_code.upper()

    if coupon_code == "SAVE10":
        discount = price * 0.10
        final_price = price - discount

    elif coupon_code == "SAVE20":
        discount = price * 0.10
        final_price = price - discount

    else:
        return "Invalid coupon code"
        
    print(f"Price after coupon discount: {final_price:.2f}")
    return final_price

def final_price(price, category, coupon_code):
    price = category_discount(price, category)
    price = apply_coupon(price, coupon_code)

    tax = price * 0.08
    final = price + tax

    print(f"Final price with (8% tax): {final:.2f}")
    return f"{final:.2f}"

price = float(input("Enter price of product: "))
category = input("Enter category: ")
coupon_code = input("Enter coupon code: ")

print(final_price(price, category, coupon_code))
```

## A hotel calculates room charges:

### Function room_charges(nights, room_type) ‚Üí rates:

    Standard = $100/night
    
    Deluxe = $200/night
    
    Suite = $400/night
    
    Function additional_services(services) ‚Üí Breakfast=$20, Spa=$50, Gym=$10
    
    Function apply_membership_discount(amount, membership_years) ‚Üí 5% discount if membership > 2 years, 10% if >5 years
    
    Function total_bill(nights, room_type, services, membership_years) ‚Üí prints detailed bill

### üì• Input Example:
    total_bill(4, "Deluxe", ["Breakfast", "Spa"], 6)

### üìù Example Output:
Room Charges: 800
Additional Services: 70
Membership Discount: 87
Total Amount: 783

```python
def room_charges(nights, room_type):
    room_type = room_type.lower()

    if room_type == "standard":
        return nights * 100
    elif room_type == "deluxe":
        return nights * 200
    elif room_type == "suite":
        return nights * 400
    else:
        return 0   

def additional_services(services):
    service_prices = {
        "breakfast": 20,
        "spa": 50,
        "gym": 10
    }

    total_service_cost = 0
    for s in services:
        s = s.lower()
        if s in service_prices:
            total_service_cost += service_prices[s]

    return total_service_cost

def apply_membership_discount(amount, membership_years):
    if membership_years > 5:
        discount = amount * 0.10
    elif membership_years > 2:
        discount = amount * 0.05
    else:
        discount = 0

    return discount

def total_bill(nights, room_type, services, membership_years):
    room_cost = room_charges(nights, room_type)
    service_cost = additional_services(services)
    
    subtotal = room_cost + service_cost
    membership_discount = apply_membership_discount(subtotal, membership_years)
    
    final_amount = subtotal - membership_discount

    print(f"Room Charges: {room_cost}")
    print(f"Additional Services: {service_cost}")
    print(f"Membership Discount: {membership_discount}")
    print(f"Total Amount: {final_amount}")

total_bill(4, "Deluxe", ["Breakfast", "Spa"], 6)
```

## A school calculates student performance:

### Each student has marks in Math, Science, English, History

    Function average_score(marks) ‚Üí calculates average
    
    Function grade(avg) ‚Üí nested function assigns grade:
    
    ‚â•90 ‚Üí A, 80‚Äì89 ‚Üí B, 70‚Äì79 ‚Üí C, 60‚Äì69 ‚Üí D, <60 ‚Üí F
    
    Function honor_roll(avg, extracurricular) ‚Üí gives "Honor Roll" if avg ‚â•85 and extracurricular ‚â•3
    
    Function student_summary(student) ‚Üí prints name, average, grade, honor roll status

### üì• Input Example:
    student = {"name":"Ali", "marks":[85,90,80,95], "extracurricular":3}
    student_summary(student)

### üìù Example Output:
    Name: Ali
    Average Score: 87.5
    Grade: B
    Honor Roll: Yes

```python
def average_score(marks):
    return sum(marks) / len(marks)

def grade(avg):

    def get_grade(a):
        if a >= 90:
            return "A"
        elif 80 <= a < 90:
            return "B"
        elif 70 <= a < 80:
            return "C"
        elif 60 <= a < 70:
            return "D"
        else:
            return "F"

    return get_grade(avg)

def honor_roll(avg, extracurricular):
    if avg >= 85 and extracurricular >= 3:
        return "Yes"
    else:
        return "No"

def student_summary(student):
    name = student["name"]
    marks = student["marks"]
    activities = student["extracurricular"]

    avg = average_score(marks)
    student_grade = grade(avg)
    honor_status = honor_roll(avg, activities)

    print(f"Name: {name}")
    print(f"Average Score: {avg}")
    print(f"Grade: {student_grade}")
    print(f"Honor Roll: {honor_status}")

student = {
    "name": "Ali",
    "marks": [85, 90, 80, 95],
    "extracurricular": 3
}

student_summary(student)
```

## A store applies layered discounts:

    Function base_discount(price) ‚Üí 5%
    
    Function category_discount(price, category) ‚Üí Electronics=10%, Clothing=8%, Grocery=3%
    
    Function coupon_discoun
    
    Function final_price(price, category, coupon) ‚Üí nested calls calculate total discounts + final price

### üì• Input Example:
    final_price(500, "Electronics", "SAVE10")

### üìù Example Output:
    Price after base discount: 475
    Price after category discount: 427.5
    Price after coupon: 417.5
    Final Price: 451.3 (with 8% tax)

```python
def base_discount(price):
    base_discount = price * 0.05
    base_price = price - base_discount
    print(f"Price after base discount: {base_price}")
    return base_price

def category_discount(price, category):
    category = category.lower()
    
    if category == "electronics":
        discount = price * 0.10
        final_price = price - discount
        
    elif category == "clothing":
        discount = price * 0.08
        final_price = price - discount

    elif category == "groccery":
        discount = price * 0.03
        final_price = price - discount

    else:
        return "Invalid Category"

    print(f"Price after category discount: {final_price}")
    return final_price

def coupon_discount(price, coupon_code):
    coupon_code = coupon_code.upper()
    
    if coupon_code == "SAVE10":
        discount = price * 0.10
        final_price = price - discount

    elif coupon_code == "SAVE20":
        if price > 200:
            discount = price * 0.20
            final_price = price - discount
        else:
            return "Order must be above 200 PKR."
    else:
        return "Invalid Coupon Code"
    
    print(f"Price after coupon discount: {final_price}")
    return final_price

def final_price(price, category, coupon_code):
    price = base_discount(price)
    price = category_discount(price, category)
    price = coupon_discount(price, coupon_code)

    tax = price * 0.08
    final = price + tax

    print(f"Final Price (after 8% tax): {final:.2f}")
    return f"{final:.2f}"

price = float(input("Enter price of product: "))
category = input("Enter category: ")
coupon_code = input("Enter coupon code: ")

print(final_price(price, category, coupon_code))
```
